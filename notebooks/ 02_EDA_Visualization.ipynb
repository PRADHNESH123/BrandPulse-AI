```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import os

# Load cleaned data
df = pd.read_csv('C:/data/BrandPulse-AI/data/processed/cleaned_tweets.csv')

# Create figures folder
os.makedirs('C:/data/BrandPulse-AI/reports/figures', exist_ok=True)

print(f"Shape: {df.shape}")
print(df.head(3))
print(df['polarity'].value_counts())
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='polarity', data=df, palette=['#FF6B6B', '#4ECDC4'])
plt.title('Sentiment Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Sentiment (0=Negative, 1=Positive)', fontsize=12)
plt.ylabel('Number of Tweets', fontsize=12)
plt.xticks([0, 1], ['Negative', 'Positive'], fontsize=11)
plt.tight_layout()
plt.savefig('C:/data/BrandPulse-AI/reports/figures/sentiment_distribution.png', dpi=150)
plt.show()
print("Saved sentiment_distribution.png ✅")
```

```python
# Get all positive tweets text
positive_text = ' '.join(df[df['polarity'] == 1]['cleaned_text'].dropna().tolist())

wordcloud_pos = WordCloud(
    width=1200,
    height=600,
    background_color='white',
    colormap='Greens',
    max_words=200,
    collocations=False
).generate(positive_text)

plt.figure(figsize=(15, 7))
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in POSITIVE Tweets ☀️',
          fontsize=18, fontweight='bold', color='green', pad=20)
plt.tight_layout()
plt.savefig('C:/data/BrandPulse-AI/reports/figures/wordcloud_positive.png', dpi=150)
plt.show()
print("Saved wordcloud_positive.png ✅")
```

```python
# Get all negative tweets text
negative_text = ' '.join(df[df['polarity'] == 0]['cleaned_text'].dropna().tolist())

wordcloud_neg = WordCloud(
    width=1200,
    height=600,
    background_color='black',
    colormap='Reds',
    max_words=200,
    collocations=False
).generate(negative_text)

plt.figure(figsize=(15, 7))
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in NEGATIVE Tweets ☁️',
          fontsize=18, fontweight='bold', color='red', pad=20)
plt.tight_layout()
plt.savefig('C:/data/BrandPulse-AI/reports/figures/wordcloud_negative.png', dpi=150)
plt.show()
print("Saved wordcloud_negative.png ✅")
```
