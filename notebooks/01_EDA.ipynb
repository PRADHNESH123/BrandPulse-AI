```python
import pandas as pd
import sys
sys.path.append('C:/data/BrandPulse-AI')
from src.preprocessing import clean_tweet

# Install tqdm first if needed
import subprocess
subprocess.run(['pip', 'install', 'tqdm', '-q'])
from tqdm import tqdm
tqdm.pandas()

cols = ['polarity', 'id', 'date', 'query', 'user', 'text']
df = pd.read_csv('C:/data/BrandPulse-AI/data/raw/training.1600000.processed.noemoticon.csv',
                 encoding='latin-1', names=cols)

df = df[['polarity', 'text']]
df['polarity'] = df['polarity'].map({0: 0, 4: 1})

print("Cleaning tweets...")
df['cleaned_text'] = df['text'].progress_apply(clean_tweet)

# Save immediately
df.to_csv('C:/data/BrandPulse-AI/data/processed/cleaned_tweets.csv', index=False)
print(f"Done! Saved! Shape: {df.shape}")
```
